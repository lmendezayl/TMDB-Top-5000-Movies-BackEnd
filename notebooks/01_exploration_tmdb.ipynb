```python
import pandas as pd
import json
```

```python
df_movies = pd.read_csv('../data/bronze/tmdb_5000_movies.csv')
df_credits = pd.read_csv('../data/bronze/tmdb_5000_credits.csv')
```

```python
df_movies.head()
```

```python
language_data = [
    {'id': language['iso_639_1'], 'name': language['name']}
    for _, row in df_movies.iterrows()
    for language in json.loads(row['spoken_languages'])
]

df_languages = pd.DataFrame(language_data).set_index('id').drop_duplicates().sort_values('id')
df_languages
```

```python
directors_data = [
    {'id': member['id'], 'name': member['name']}
    for _, row in df_credits.iterrows()
    for member in json.loads(row['crew'])
    if member['job'] == 'Director'
]

genres_data = [
    {'id': genre['id'], 'name': genre['name']}
    for _, row in df_movies.iterrows()
    for genre in json.loads(row['genres'])
]



df_directors = pd.DataFrame(directors_data).set_index('id').drop_duplicates().sort_values('id')
df_genres = pd.DataFrame(genres_data).set_index('id').drop_duplicates().sort_values('id')
```

```python
genres_data = [
    {'id': genre['id'], 'name': genre['name']}
    for i, row in df_movies.iterrows()
    for genre in json.loads(row['genres'])
]

df_genres = pd.DataFrame(genres_data).set_index('id').drop_duplicates().sort_values('id')
df_genres
```
